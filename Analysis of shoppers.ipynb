```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import preprocessing
```

```python
df=pd.read_csv('Shoppers.csv')
```

```python
df['X_MoB']= preprocessing.scale(df['MoB'])
df['X_MaxBalance']= preprocessing.scale(df['MaxBalance'])
df['X_PchActiveMths']= preprocessing.scale(df['PchActiveMths'])
df['X_FinCharges']= preprocessing.scale(df['FinCharges'])
df['X_OnlineTxn']= preprocessing.scale(df['OnlineTxn'])
df['X_PaytoBalance']= preprocessing.scale(df['PaytoBalance'])
df['X_ActiveCards']= preprocessing.scale(df['ActiveCards'])
```

```python
df.head()
```

```python
df.head()
```

```python
df.drop(['x_mob'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['X_OfflinePay']= preprocessing.scale(df['OfflinePay'])
```

```python
df.drop(['OfflinePay'],axis=1,inplace=True)
df.head()
```

```python
sns.pairplot(df[['X_MoB','X_MaxBalance','X_PchActiveMths','X_FinCharges','X_OnlineTxn','X_PaytoBalance','X_ActiveCards','X_OfflinePay']])
```

```python
sns.pairplot(df[['X_MoB','X_OnlineTxn','X_OfflinePay']])
```

```python
from sklearn.cluster import KMeans
```

```python
k_range=range(1,12)
clusters= []
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[['X_MoB','X_OfflinePay']])
    clusters.append(km.inertia_)
```

```python
clusters
```

```python
plt.xlabel('Number of clusters (k)')
plt.ylabel('Number of squared error')
plt.plot(k_range,clusters)
```

```python
kmeans_cluster=KMeans(n_clusters=4)
```

```python
kmeans=kmeans_cluster.fit(df[['X_MoB','X_OfflinePay']])
```

```python
kmeans.cluster_centers_
```

```python
df['mob_clusters']=kmeans.labels_
```

```python
df
```

```python
df['mob_clusters'].value_counts()
```

```python
sns.scatterplot(x='X_MoB',y='X_OfflinePay',hue='mob_clusters',data=df)
```
